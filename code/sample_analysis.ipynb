# How much do people really like cats?
### Summary Results
On average there seems to be a significant difference between participants' preference for 3 different cat breeds *F*(2,561) = 1.26, *p* < 0.001.  
However, pairwise differences don't appear appear to be significant, all *ts* < 1.6, all *ps* > 0.10.

## Import some libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
import statsmodels.api as sm
import statsmodels.formula.api as smf
%matplotlib inline
```

## Load in data

```python
df = pd.read_csv('sample_data.csv')
```

## Visualize data

```python
sns.set_context("poster")
_,ax=plt.subplots(1,1,figsize=(10,6))
ax = sns.barplot(x='Cat Breed',y='Liking',data=df,ax=ax);
ax.set_ylabel('Average Liking');
```

## Fit a model

```python
model = smf.ols(formula='Liking ~ Cat_Breed',data=df).fit()
table = sm.stats.anova_lm(model, typ=3)
```

### Print ANOVA table

```python
print(table)
```

### Print parameter results

```python
model.summary()
```
